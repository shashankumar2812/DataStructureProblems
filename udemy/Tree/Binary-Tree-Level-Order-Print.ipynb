# Tree Level Order Print

Given a binary tree of integers, print it in level order. The output will contain space between the numbers in the same level, and new line between different levels. For example, if the tree is:




# Solution

```python
class Node: 
    def __init__(self, val=None):
        self.val=val
        self.left=None
        self.right=None      
```

```python
def tree_level_print(tree):
    
    if not tree:
        return
    queue=[tree]
    current_count,next_count=1,0
    while(queue):
        
        current_node=queue.pop(0)
        current_count-=1
        print(current_node.val,end=" ")
        if current_node.left:
            queue.append(current_node.left)
            next_count+=1
        if current_node.right:
            queue.append(current_node.right)
            next_count+=1
        if current_count==0:
            print("\n")
            current_count, next_count=next_count, current_count   
```

```python
root=Node(1)
root.left=Node(2)
root.right=Node(3)
root.left.left=Node(4)
root.right.left=Node(5)
root.right.right=Node(6)
```

```python
tree_level_print(root)
```

```python
from collections import deque
```

```python
def print_tree_level(root):
    current_depth=1
    queue=deque()
    queue.append((root,1))
    while queue:
       
        current_node, d=queue.popleft()
        if d== current_depth:
            print(current_node.val, end=" ")
        else:
            print('\n')
            print(current_node.val, end=" ")
            current_depth=d
        if current_node.left:
            queue.append((current_node.left,d+1))
        if current_node.right:
            queue.append((current_node.right,d+1))   
```

```python
print_tree_level(root)
```
